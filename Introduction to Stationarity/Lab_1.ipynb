```python
# 7 show that the features are stationary

# Function to get cumulative sum at each element stage
def summation(feature):
    dfl = feature
    summation = []
    total = 0
    for i in dfl:
        total += i
        summation.append(total)
    return summation
    
```

```python
# Function to get the average at each element stage
def means(feature):
    agg = feature
    means = []
    for i in list(range(1,101)):
        x = agg[i-1]/i
        means.append(x)
    return means
```

```python
# Functionto plot the change in statstic with each time step
def plots_mean(stat,title,feature):
    x = range(1,101)
    line = plt.plot(x,stat,label = str(feature))
    plt.xlabel("Quarter")
    plt.ylabel("Mean (in 1000s of $)")
    plt.legend(loc="upper right")
    plt.title(str(title))
    plt.ylim(0,1.5*max(stat))
    return plt.show
```

```python
# Functionto plot the change in statstic with each time step
def plots_var(stat,title,feature):
    x = range(2,100)
    plt.plot(x,stat)
    line = plt.plot(x,stat,label = str(feature))
    plt.xlabel("Quarter")
    plt.ylabel("Variance (in 1000s of $)")
    plt.legend(loc="upper right")
    plt.title(str(title))
    plt.ylim(0,1.5*max(stat))
    return plt.show
```

```python
# #9 Perform a an ADF test to check if the Sales, Adbudget and GDP are stationary or not? 
# Does your answer for this question reinforce your observations in the previous steps?
```

```python
def ADF_Cal(x):
    result = adfuller(x)
    print("ADF Statistic: %f" %result[0])
    print("p-value: %f" %result[1])
    print("Critical Values:")
    for key, value in result[4].items():
        print('\t%s: %3f' % (key,value))
```

```python
ADF_Cal(df.Sales)
```

```python
# Since the critical value is less than the 5% threshold, the null hypothesis that the data has a unit root and is non-stationary
# is rejected in favor of the alternative hypotheis that the data does not have a unit root and is stationary.
```

```python
ADF_Cal(df.AdBudget)
```

```python
# Since the critical value is less than the 5% threshold, the null hypothesis that the data has a unit root and is non-stationary
# is rejected in favor of the alternative hypotheis that the data does not have a unit root and is stationary.
```
